```python
import pandas as pd
df = pd.read_csv("bos2013_demo_glvt_kmean15.csv", sep=";")
df = df[df["premium_basis"] == 'Premium Segment']

df = df[['kmean15', 'gender', 'marstaall']]
df = df.sort_values(by='kmean15')
df
```

```python
sort_list = ['gender', 'marstaall']
```

```python
data = {
    'kmean15': []
}
for i in set(df.kmean15.tolist()):
    data['kmean15'].append(i)
for column_name in sort_list:
    tmp_set = set(df[column_name].tolist())
    types = [i for i in tmp_set if i != ' ' if i != '']
    for i in types:
        data[i] = []
    for i in set(df.kmean15.tolist()):
        values = df[column_name][df.kmean15 == i].value_counts()
        for j in types:
            try:
                tmp_value = values[j]
            except KeyError as e:
                tmp_value = 0
            data[j].append(tmp_value)

            
df_data = pd.DataFrame(data=data)
df_data
```
